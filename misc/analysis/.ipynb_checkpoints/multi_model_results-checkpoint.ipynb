This notebook is used to analyze the results of the multi-model experiments.

```python
import sys
sys.path.append("../")

from pathlib import Path
import json
from src.ensembles import MODEL_GROUPS
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import numpy as np
from typing import List
```

```python
# CONSTANTS
RESULTS_DIR = Path("../smoothie_data/multi_model_results")
TASK2METRIC = {
    "cnn_dailymail": "rouge2",
    "definition_extraction": "definition_extraction_acc",
    "e2e_nlg": "rouge2",
    "squad": "squad_acc",
    "trivia_qa": "trivia_qa_acc",
    "web_nlg": "rouge2",
    "xsum": "rouge2",
}
SINGLE_TASK_DATASETS = ["squad", "trivia_qa", "definition_extraction", "cnn_dailymail",  "xsum", "e2e_nlg", "web_nlg"]

# 
```

```python
fpath = RESULTS_DIR / "squad" / "ensemble_40" / "scores.json"
scores = json.loads(fpath.read_text())["squad_acc"]
scores
```

```python
def plot_single_task_win_rates(datasets: List[str], method_a: str, method_b: str):
    """
    """ 
    task_avg_diffs = [] # Track average difference between smoothie and random
    for dataset in datasets:
        diffs = []
        for model_group in MODEL_GROUPS:
            scores_fpath = RESULTS_DIR / dataset/ model_group / "scores.json"
            scores = json.loads(scores_fpath.read_text())[TASK2METRIC[dataset]]
            diffs.append(int(scores["smoothie"][method_a] > scores[method_b]))
        task_avg_diffs.append(np.mean(diffs))

     # Set the style
    plt.style.use('ggplot')
    
    # Create a custom colormap
    colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']
    n_bins = len(datasets)
    cmap = LinearSegmentedColormap.from_list('custom', colors, N=n_bins)
    
    # Create bar plot
    plt.figure(figsize=(14, 8))
    bars = plt.bar(datasets, task_avg_diffs, width=0.6, edgecolor='black', linewidth=1.5, color=cmap(np.arange(n_bins)/n_bins))
    
    # Customize the plot
    plt.title(f"Win Rate of {method_a} over {method_b} by Dataset", fontsize=20, pad=20)
    plt.xlabel("Dataset", fontsize=14, labelpad=10)
    plt.ylabel("Win Rate", fontsize=14, labelpad=10)
    plt.ylim(0, 1)  # Set y-axis limits from 0 to 1
    plt.xticks(rotation=45, ha='right', fontsize=12)
    plt.yticks(fontsize=12)
    
    # Add a subtle grid
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    
    # Add value labels on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'{height:.2f}',
                 ha='center', va='bottom', fontsize=11, fontweight='bold')
    
    # Add a legend to explain what the win rate means
    plt.legend(['Win rate (higher is better for ' + method_a + ')'], loc='upper right', fontsize=12)
    
    plt.tight_layout()  # Adjust layout to prevent clipping of labels
    plt.show()
    return task_avg_diffs
    
            
    

    
```

```python
plot_single_task_win_rates(
    datasets=SINGLE_TASK_DATASETS,
    method_a="smoothie_sample_independent_train_time_all-mpnet-base-v2",
    method_b="pick_random"
)
```
